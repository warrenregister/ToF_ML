```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('./data/processed_cas.csv')
for column in df.columns:
    try:
        df[column] = df[column].apply(str.strip)
    except:
        pass
df.drop('Unnamed: 0', inplace=True, axis=1)
```

```python
def get_data_points(channels):
    '''
    Convert sequence data which represents channels with -#s as the # 
    of channels with 0s in it in a row to full channel data for graphing
    by converting -#s to that number of 0s.
    '''
    points = []
    nums = channels.split()
    for num in nums:
        num = int(num)
        if num >= 0:
            points.append(num)
        else:
            for i in range(abs(num)):
                points.append(0)
    return points[1:]
df['channels'] = df['sequence'].apply(get_data_points)
```

```python
plt.plot(df['channels'][4])
```

```python
def mass_formula(channel, spec_bin_size,start_time,  mass_over_time, mass_offset):
        '''
        Apply formula for calculating mass at a channel.
        '''
        return ((channel * .001 * spec_bin_size + start_time) * mass_over_time + mass_offset)**2
```

```python
def generate_calibrated_data():
    '''
    Applies mass_formula to every row in dataset to allow
    calibrated graphs to be generated.
    '''
    spectrums = []
    for row in df.iterrows():
        points = []
        spec = row[1]['SpecBinSize']
        m_over_t = row[1]['Mass/Time']
        m_offset = row[1]['MassOffset']
        time = row[1]['StartFlightTime']
        for num in range(len(row[1]['channels'])):
            points.append(mass_formula(num, spec, time, m_over_t, m_offset))
        spectrums.append(points)
    return spectrums
```

```python
x_axis = generate_calibrated_data()
```

```python
df['calibrated_x_axis'] = x_axis
```
