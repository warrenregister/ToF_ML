```python
def convert_sequence(seq):
    channels = []
    for char in seq.split()[1:]:
        num = int(char)
        if num < 0:
            channels = channels + [0 for x in range(abs(num))]
        else:
            channels.append(num)
    return pd.Series(channels)
```

```python
df['Full Channels'] = df['sequence'].apply(convert_sequence)
```
